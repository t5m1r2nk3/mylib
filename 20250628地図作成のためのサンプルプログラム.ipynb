```python
#%地図データ取得
import gcsfs
import geopandas as gpd

#!gsutil ls gs://map_niwa/geojson/

df_city = gpd.read_file('gs://map_niwa/geojson/hokkaido_city.geojson')
df_sinkou = gpd.read_file('gs://map_niwa/geojson/hokkaido_sinkoukyoku.geojson')
df_hoppou = gpd.read_file('gs://map_niwa/geojson/hokkaido_hoppou.geojson')
df_city_point=gpd.read_file('gs://map_niwa/geojson/hokkaido_city_point.geojson')


```

```python
#データ準備
import pandas as pd
df=pd.read_excel('/content/2017_MunicipalFoodWaste.xlsx')
col=pd.DataFrame(df.columns)
col
df=df.iloc[:,[2,3]]
#[市区町村名,3	生活系ごみ収集量_t]
df.head()

#緯度経度情報追加
df=pd.merge(df,df_city_point,left_on='市区町村名',right_on='NO3_004',how='left')


```

```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable
import matplotlib.colors
import matplotlib.pyplot as plt

#!pip install japanize_matplotlib
import japanize_matplotlib

#%%
# カラーマップを定義します

############
#円にしたい列名
volume='生活系ごみ収集量_t'
#円の大きさ調整(rがおおきいほど小さくなる)

r=200
#円の色
c='m'
############
HANREI_vmax=df[volume].max()

fig = plt.figure(figsize=(8,6),dpi=150)
ax = fig.add_subplot(111)
ax.set_title(volume, fontsize=14)
# legend_kwds = dict(bbox_to_anchor=(0.65, 0.65), loc='upper left', borderaxespad=0,labels=s_labels,#bbox_to_anchor=(0.65, 0.4)
#                    title=legend_title,fontsize=14,frameon = False,title_fontsize=14,fmt='{:,.0f}')
# scheme="quantiles"カテゴリー
#市町村境界線追加
df_city.plot(ax=ax, color='none', edgecolor='k', linewidth=0.1)
#北方領土追加
df_hoppou.plot(column="N03_001",legend=False,ax=ax,  color='white')
#振興局境界線追加
df_sinkou.plot(ax=ax, color='none', edgecolor='k', linewidth=0.8)#linewidth=0.2,0.5

############################################
#その１ 量を円の大きさで表現する(どちらかコメントアウト)
############################################
# =============================================================================
# norm = Normalize(vmin=df[volume].min()/r, vmax=df[volume].max()/r)#その１その２どちらも必要
# for idx, row in df.iterrows():
#     #円の大きさで表す
#     color = cmap(norm(row[volume]))
#     plt.scatter(row.geometry.x, row.geometry.y, s=row[volume]/r, alpha=0.5,color=c)


# # 凡例を追加します
# #凡例は２種類の方法。「A」は単純按分。「B」は任意で。どちらかをコメントアウト

# ################「A」方法##########################
# # sizes = list(range(5))
# # labels= list(range(5))
# # for i in range(5):
# #     labels[i]=str(int(HANREI_vmax*(0.2+0.2*i)))
# #     sizes[i]=HANREI_vmax/r*(0.2+0.2*i)
# ####################################################

# ################「B」方法##########################
# sizes = [10000, 50000, 100000, 200000,400000]  # 円の大きさの範囲
# labels = [f'{str(i)}'for i in sizes]  # 対応する数量の範囲
# for size, label in zip(sizes, labels):
#     plt.scatter([], [], c=c, alpha=0.5, s=size/r, label=label)

# plt.legend(scatterpoints=1, frameon=False, labelspacing=2, title=volume,bbox_to_anchor=(1, 0.8))
# ###################################################

# ax.axis("off")

# volume_label=volume.replace('（kg/d）','')

# plt.savefig('円の大きさで表現.png', format="png", dpi=300)
# =============================================================================

############################################
#その２　円の色で表現する(どちらかコメントアウト)
############################################
# =============================================================================
#norm = Normalize(vmin=df[volume].min(), vmax=df[volume].max())#その１その２どちらも必要
norm = Normalize(vmin=0, vmax=10000)#その１その２どちらも必要
cmap=matplotlib.colormaps['viridis']#その２　円の色で表現するとき必要

for idx, row in df.iterrows():
    #円の色で表す
    color = cmap(norm(row[volume]))
    plt.scatter(row.geometry.x, row.geometry.y, s=20, alpha=1,color=color)

# カラーバーを追加します
sm = ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar=plt.colorbar(sm, label=volume,shrink=0.4,ax=plt.gca())#shrinkはカラーバーの大きさ
cbar.ax.set_position([0.65, 0.25, 0.4, 0.4])# [左端の位置 (left)、下端の位置 (bottom)、幅 (width:倍率)、高さ (height：倍率)]


plt.legend(scatterpoints=1, frameon=False, labelspacing=2, title='収集量凡例',bbox_to_anchor=(1, 0))

ax.axis("off")
plt.savefig('円の色で大きさ表現.png', format="png", dpi=300)
# =============================================================================
```

#緯度経度テスト（施設の違いを色で表現：サンプル「生ごみ発生量」）

```python
#データの準備

#データ取得
import gcsfs
import pandas as pd

df=pd.read_excel('gs://sample_data_niwa/生ごみ発生量.xlsx')

#緯度経度情報追加
df=pd.merge(df,df_city_point,left_on='市区町村名',right_on='NO3_004',how='left')

df=df.iloc[:,[1,2,6]]
df['方法'].unique().tolist()
#['焼却（発生）', 'その他（発生）', '堆肥化（発生）', '焼却（処分）']
#'焼却（処分）'を削除
df = df[df['方法'] != '焼却（処分）']
```

```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable
import matplotlib.colors
import matplotlib.pyplot as plt

#!pip install japanize_matplotlib
import japanize_matplotlib


fig = plt.figure(figsize=(8,6),dpi=150)
ax = fig.add_subplot(111)
ax.set_title(volume, fontsize=14)

#市町村境界線追加
df_city.plot(ax=ax, color='none', edgecolor='k', linewidth=0.1)
#北方領土追加
df_hoppou.plot(column="N03_001",legend=False,ax=ax,  color='white')
#振興局境界線追加
df_sinkou.plot(ax=ax, color='none', edgecolor='k', linewidth=0.8)#linewidth=0.2,0.5

# 施設ごとに色分け
# use=['焼却（発生）', 'その他（発生）', '堆肥化（発生）', '焼却（処分）']
# color_u=['k','g','b','r']

# 施設ごとの方法と色の対応を辞書で定義
method_colors = {
    '焼却（発生）': 'k',  # 黒
    'その他（発生）': 'g',  # 緑
    '堆肥化（発生）': 'b',  # 青
    '焼却（処分）': 'r'   # 赤
}

# プロット本体
for idx, row in df.iterrows():
    color = method_colors.get(row['方法'], 'gray')  # 未定義の方法には灰色
    ax.scatter(row.geometry.x, row.geometry.y, marker='o', alpha=1, color=color)

# 凡例を追加します
# use=['焼却（発生）', 'その他（発生）', '堆肥化（発生）', '焼却（処分）']
# color_u=['k','g','b','r']

# for i, j in zip(use,color_u):
#     plt.scatter([], [], c=j, alpha=0.5, s=20, label=i)

for method, color in method_colors.items():
    ax.scatter([], [], c=color, alpha=1, s=20, label=method)


plt.legend(frameon=False, title='凡例',loc='lower right',bbox_to_anchor=(1, 0.2))

#plt.legend(handles=legend_patches,frameon=False, title='凡例',loc='lower right')

ax.axis("off")
plt.savefig('生ごみ処理.png', format="png", dpi=300)



```
